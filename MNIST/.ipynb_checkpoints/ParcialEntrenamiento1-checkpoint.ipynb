# Importamos las librerias

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.utils import to_categorical
from keras import models
from keras import layers
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
from sklearn.model_selection import train_test_split
from pandas import read_csv
from pycaret.classification import setup
from pycaret.classification import compare_models
```

## Cargamos la data y la separamos en Train y Test

```python
X=np.load('X_MNIST_60_mil.npy', allow_pickle=True)
Y=np.load('y_MNIST_60_mil.npy', allow_pickle=True)
Xfinal=np.load('X_MNIST_10_mil.npy', allow_pickle=True)
```

```python
train_imagen, test_imagen, train_numero, test_numero = train_test_split(X, Y, test_size=0.30, random_state=123)
```

### Muestro el tama√±o

```python
print(train_imagen.shape)
print(test_imagen.shape)
print(train_numero.shape)
print(test_numero.shape)
```

### Graficamos algunos valores

```python
nn = 5
for i, num in enumerate([random.randint(1, train_imagen.shape[0]) for _ in range(nn)]):
    plt.subplot(1, nn, i+1)
    plt.xlabel(train_numero[num])
    plt.imshow(train_imagen[num].reshape(28, 28))
```

--------------------------

# Creamos la red neuronal

```python
red = models.Sequential()
red.add(layers.Dense(512, activation='relu'))
red.add(layers.Dense(400, activation='relu'))
red.add(layers.Dense(320, activation='relu'))
red.add(layers.Dense(250, activation='relu'))
red.add(layers.Dense(200, activation='relu'))
red.add(layers.Dense(180, activation='relu'))
red.add(layers.Dense(150, activation='relu'))
red.add(layers.Dense(100, activation='relu'))
red.add(layers.Dense(50, activation='relu'))
red.add(layers.Dense(20, activation='relu'))
red.add(layers.Dense(10, activation='softmax'))
```

## La compilamos

```python
red.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
```

## Normalizamos los valores

```python
train_imagen = train_imagen.astype('float32') / 255
test_imagen = test_imagen.astype('float32') / 255
```

```python
train_numero_cat = to_categorical(train_numero)
test_numero_cat = to_categorical(test_numero)
```

## Entrenamos el modelo 

```python
history = red.fit(train_imagen, train_numero_cat, epochs=25, batch_size=128, validation_split=0.2, shuffle=True)
```

## Revisamos el score

```python
test_loss, test_acc = red.evaluate(test_imagen, test_numero_cat)
print('test_acc: ', test_acc)
```

### Datos finales del modelo

```python
red.summary()
```

### Vemos la curva entre los Accuracy y los Loss de Train y Validation

```python
plt.figure(figsize=(14,8))
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'val'], loc='upper left')
plt.show()
```

```python
plt.figure(figsize=(14,8))
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['loss', 'val_loss'], loc='upper left')
plt.show()
```
